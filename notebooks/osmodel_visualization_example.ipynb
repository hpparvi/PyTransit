# OblateStarModel visualisation example 

```python
%pylab inline
```

```python
from pytransit import OblateStarModel
```

```python
k = 0.1
p, a, e, w = 1.5, 4.5, 0.0, 0.0
b = 0.0
alpha = 0.1 * pi

ldc = array([0.3, 0.1])
tpole = 6400.
istar = 0.15 * pi
beta = 0.3
rperiod = 0.3
rho = 1.5

m = OblateStarModel(sres=125, pres=15, tres=60)
```

```python
m.visualize(k, p, rho, b, e, w, alpha, rperiod, tpole, istar, beta, ldc)
```
